```python
import numpy as np
import pandas as pd
import anndata as ad

gene_symbols = adata.var['gene_symbol'].values
timepoints = list(adata.varm.keys())

all_embeddings = []
obs_names = []
obs_timepoints = []
obs_genes = []

for tp in timepoints:
    embedding = adata.varm[tp]  
    all_embeddings.append(embedding)
    
    obs_names.extend([f"{g}_{tp}" for g in gene_symbols])
    obs_genes.extend(gene_symbols)
    obs_timepoints.extend([tp] * len(gene_symbols))

X = np.vstack(all_embeddings)

obs = pd.DataFrame({
    "gene_symbol": obs_genes,
    "timepoint": obs_timepoints
}, index=obs_names)

new_adata = ad.AnnData(X=X, obs=obs)
```

```python
## Get hint from predefined gene sets
import gseapy as gp

cluster_results = {}

for cluster in new_adata.obs['r1'].unique():
    gene_list = new_adata.obs.loc[new_adata.obs['r1'] == cluster, 'gene_symbol'].tolist()
    
    enr = gp.enrichr(
        gene_list=gene_list,
        gene_sets=['Reactome_Pathways_2024', 'GO_Biological_Process_2025'],
        organism='Human',         
        outdir=None,                
        cutoff=0.05 ,
        background=background_genes
    )
    
    top_terms = enr.results.sort_values('Adjusted P-value').head(10)
    cluster_results[f"Cluster_{cluster}"] = top_terms

combined = pd.concat(cluster_results, names=['Cluster', 'Index'])
combined.reset_index(inplace=True)
```

```python
import pandas as pd
import scanpy as sc

# Assuming new_adata is your AnnData object
# Create a contingency table (cross-tabulation) of clusters vs timepoints
contingency_table = pd.crosstab(
    new_adata.obs['r1'], 
    new_adata.obs['timepoint']
)

# To find timepoint-specific clusters, look for rows (clusters) that have counts in only one column (timepoint)
timepoint_specific_clusters = contingency_table[(contingency_table > 0).sum(axis=1) == 1]

print("Timepoint-specific clusters:")
print(timepoint_specific_clusters)

# Optional: Add proportions to better understand the specificity
proportion_table = contingency_table.div(contingency_table.sum(axis=0), axis=1)
print("\nProportions per timepoint:")
print(proportion_table)
```

```python
import umap
from adjustText import adjust_text
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

highlight_dict = {
    'Cytokine Signaling': ["IL1B", "PTX3","IL7R","CD14","IL32","CXCL8","CX3CR1","XCL1","CCL2"],

}
umap_df = pd.DataFrame(new_adata.obsm['X_umap'], columns=['UMAP1', 'UMAP2'], index=new_adata.obs_names)
umap_df['MoA'] = new_adata.obs['MoA']
umap_df['gene'] = new_adata.obs['gene_symbol']  

umap_model = umap.UMAP(min_dist=0.3, spread=3.0, random_state=42)
umap_embedding = umap_model.fit_transform(umap_df[['UMAP1', 'UMAP2']])

custom_colors = {
    'Cytokine Signaling': '#d9db05',  
    'others': '#939da3',  
}

moa_colors = umap_df['MoA'].map(custom_colors).fillna('#000000')  

plt.figure(figsize=(20, 14))  
sns.scatterplot(
    data=umap_df, x='UMAP1', y='UMAP2', hue=umap_df['MoA'],  
    palette=custom_colors, linewidth=0, alpha=0.75, s=30  
)

plt.title('UMAP with Annotated Genes (Customized)', fontsize=30)
plt.xlabel('UMAP1', fontsize=24)
plt.ylabel('UMAP2', fontsize=24)

texts = []
for moa, genes in highlight_dict.items():
    for gene in genes:
        mask = (umap_df['gene'] == gene) & (umap_df['MoA'] == moa)
        if mask.any():
            row = umap_df[mask].iloc[0]
            x, y = row['UMAP1'], row['UMAP2']
            
            if gene == "IL1RN":
                jitter_x = x - 1.0  #
            else:
                jitter_x = x + np.random.uniform(-0.5, 0.5)  
            jitter_y = y + np.random.uniform(-0.5, 0.5)  

            texts.append(
                plt.text(
                    jitter_x, jitter_y, gene,
                    fontsize=18, fontweight='bold',  
                    ha='center', va='center'
                )
            )

adjust_text(
    texts,
    expand_points=(15, 15),           
    expand_text=(10, 10),            
    force_points=5.0,             
    force_text=5.0,                 
    pad=10.0,                      
    only_move={'points': 'xy', 'text': 'xy'},
    autoalign='xy',
    lim=1500                       
)

handles, labels = plt.gca().get_legend_handles_labels() 
plt.legend(
    handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=custom_colors[label], markersize=12) for label in custom_colors],
    labels=list(custom_colors.keys()), 
    bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=20
)

plt.tight_layout()
plt.grid(False)

plt.savefig('umap_customized_with_IL1RN_shifted.png', dpi=600)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

counts = (
    new_adata.obs
    .groupby(['r1', 'timepoint'])
    .size()
    .reset_index(name='count')
)

pivot_df = counts.pivot(index='r1', columns='timepoint', values='count').fillna(0)

pivot_df_normalized = pivot_df.div(pivot_df.sum(axis=1), axis=0)

pivot_df_normalized.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

plt.title('Timepoint Composition per Cluster (r1)')
plt.ylabel('Proportion of Genes')
plt.xlabel('Cluster (r1)')
plt.legend(title='Timepoint', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

intersection_counts = {}

clusters = new_adata.obs['r1'].unique()

for cluster in clusters:
    sub_df = new_adata.obs[new_adata.obs['r1'] == cluster]
    
    gene_sets = sub_df.groupby('timepoint')['gene_symbol'].apply(set)
    
    # Only proceed if all 3 timepoints are present
    if len(gene_sets) == 3:
        common_genes = set.intersection(*gene_sets)
        intersection_counts[cluster] = len(common_genes)
    else:
        intersection_counts[cluster] = 0  # or skip, if you prefer

intersection_df = pd.DataFrame.from_dict(intersection_counts, orient='index', columns=['intersection_size'])
intersection_df = intersection_df.sort_values(by='intersection_size', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=intersection_df.index.astype(str), y='intersection_size', data=intersection_df, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Cluster (r1)')
plt.ylabel('Number of Genes Shared Across All Timepoints')
plt.title('Shared Genes Across Timepoints per Cluster')
plt.tight_layout()
plt.show()
```
