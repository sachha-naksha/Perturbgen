```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
def leiden_silhouette_analysis(new_adata, knn_values, resolution_values):
    results = []
    for knn in knn_values:
        sc.pp.neighbors(new_adata, n_neighbors=knn, use_rep="X")
        for resolution in resolution_values:
            sc.tl.leiden(new_adata, resolution=resolution, key_added=f'leiden_{knn}_{resolution}')
            labels = new_adata.obs[f'leiden_{knn}_{resolution}']
            
            sil_score = silhouette_score(new_adata.obsm['X_pca'], labels)
            results.append({'knn': knn, 'resolution': resolution, 'silhouette_score': sil_score})
    
    return pd.DataFrame(results)

knn_values = [5, 10, 15, 20]  
resolution_values = [0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45, 0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,0.1]

sc.tl.pca(new_adata)

results_df = leiden_silhouette_analysis(new_adata, knn_values, resolution_values)
```

```python
best_params = results_df.loc[results_df['silhouette_score'].idxmax(), ['knn', 'resolution']]
best_params
```

```python
plt.figure(figsize=(12, 6))
bar_width = 0.2
x = np.arange(len(resolution_values))  

for i, knn in enumerate(knn_values):
    subset = results_df[results_df['knn'] == knn]
    plt.bar(
        x + i * bar_width, 
        subset['silhouette_score'], 
        width=bar_width, 
        label=f"k={knn}"
    )

plt.xticks(x + bar_width * (len(knn_values) - 1) / 2, [f"R={r}" for r in resolution_values])
plt.xlabel("Resolution")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different knn and Resolution Combinations")
plt.legend(title="knn")
plt.tight_layout()
plt.savefig('./knn_leiden.png',dpi=600)
plt.show()
```
